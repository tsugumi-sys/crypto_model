```python
import pandas as pd
import sys

sys.path.append("..")
from common.constants import DATAFOLDER, ASSET_INFO
```

```python
data_folder = DATAFOLDER.raw_data_root_path
asset_info = ASSET_INFO.asset_info
print(asset_info)
btcusdt = list(asset_info.keys())[0]
import os
data_folder = os.path.join(data_folder, btcusdt)

file_bytes = 0
csv_files = []
for csv_f in os.listdir(data_folder):
    csv_files.append(os.path.join(data_folder, csv_f))
    file_bytes += os.path.getsize(os.path.join(data_folder, csv_f))

print(file_bytes // 10**6, "MB")
```
