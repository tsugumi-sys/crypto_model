```python
import pandas as pd
import sys

sys.path.append("..")
from common.constants import DATAFOLDER, ASSET_INFO
```

```python
data_folder = DATAFOLDER.raw_data_root_path
asset_info = ASSET_INFO.asset_info
print(asset_info)
btcusdt = list(asset_info.keys())[0]
import os
data_folder = os.path.join(data_folder, btcusdt)

file_bytes = 0
csv_files = []
for csv_f in os.listdir(data_folder):
    csv_files.append(os.path.join(data_folder, csv_f))
    file_bytes += os.path.getsize(os.path.join(data_folder, csv_f))

print(file_bytes // 10**6, "MB")
```

```python
sample_csv = csv_f

_columns = ["OpenTime", "Open", "High", "Low", "Close", "Volume", "CloseTime", "QuoteAssetVolume", "NumberOfTrades", "TakerBuyBaseAssetVolume", "TakerBuyQuoteAssetVolume", "Ignore"]
df = pd.read_csv(os.path.join(data_folder, sample_csv), names=_columns, index_col="OpenTime")
print(df.dtypes)
print(df.index.dtype)
df.head()
```

```python
df.index = df.index // 1000
print(df.index.dtype)
print(df.index[0], df.index[-1])
df.head()
```

```python
df = df.sort_index()
df = df.reindex(range(df.index[0], df.index[-1] + 900, 900))
print(df.index[0], df.index[-1])
df.isna().sum().sum()
```

```python
csv_files.sort()

sample1 = csv_files[0]
sample2 = csv_files[1]
print(sample1, sample2)
```

```python
df1 = pd.read_csv(sample1, names=_columns, index_col="OpenTime")
df2 = pd.read_csv(sample2, names=_columns, index_col="OpenTime")
df1.index = df1.index // 1000
df2.index = df2.index // 1000

print(df1.index[-1], df2.index[0])
print(df1.index[-1] + 900)
assert df1.index[-1] + 900 == df2.index[0]
```

```python
df = pd.DataFrame(columns=_columns)
df = df.set_index("OpenTime")

def fillna(df: pd.DataFrame, step: int = 900):
    df = df.sort_index()
    df.index = df.index // 1000  # Convert milisecond to second.
    df = df.reindex(range(df.index[0], df.index[-1] + step, step), method="pad")
    return df

df1 = fillna(df1)
df = df.append(df1)
df.head()
```
