```python
import pandas as pd
import os

data_path = "../data/preprocess/cpcv_fold/Bitcoin/"
folds = [f"fold_{i}" for i in range(5)]
senarios = [f"senario_{i}" for i in range(5)]
types = ["test", "train", "valid"]

for fold in folds:
    for senario in senarios:
        for _type in types:
            file_dir = os.path.join(
                data_path,
                fold,
                senario,
                _type
            )
            for file in os.listdir(file_dir):
                df = pd.read_parquet(
                    os.path.join(file_dir, file),
                    engine="pyarrow",
                )
                assert len(df.index) > 0
                assert df.isna().sum().sum() == 0
```
